```python
import simulation as sk
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize.minpack import curve_fit
```

```python
def verlet_simulation(step, steps, sigma, epsilon, mass, temp, desired_temp, box_size, relax_steps):
    dm_time = step * (epsilon/(mass*sigma**2))**(1/2)
    square = sk.fcc_lattice(3,3,3,1)
    vels = sk.init_velocity(len(square), temp, mass)
    return sk.vervlet_evolution(pos=square, 
                                vel=vels, 
                                time_step=dm_time, 
                                steps=steps, 
                                mass=mass, 
                                sigma=sigma, 
                                epsilon=epsilon, 
                                size=box_size, 
                                temp=temp,
                                relaxation_steps=relax_steps, 
                                desired_temp=desired_temp)
```

```python
sigma, epsilon, mass = sk.argon_parameters()
t_step = 1e-15
temp = 100
steps = 10000
relax_steps = 10
```

```python
size = 7
desired_T = 3

gas_data, gas_energy, gas_observables = verlet_simulation(step=t_step,
                                                           steps=steps,
                                                           sigma=sigma,
                                                           epsilon=epsilon,
                                                           mass=mass,
                                                           temp=temp,
                                                           desired_temp=desired_T,
                                                           box_size=size,
                                                           relax_steps=relax_steps)
```

```python
size = 5.05
desired_T = 1

liquid_data, liquid_energy, liquid_observables = verlet_simulation(step=t_step,
                                                           steps=steps,
                                                           sigma=sigma,
                                                           epsilon=epsilon,
                                                           mass=mass,
                                                           temp=temp,
                                                           desired_temp=desired_T,
                                                           box_size=size,
                                                           relax_steps=relax_steps)
```

```python
size = 4.45
desired_T = 0.5

solid_data, solid_energy, solid_observables = verlet_simulation(step=t_step,
                                                           steps=steps,
                                                           sigma=sigma,
                                                           epsilon=epsilon,
                                                           mass=mass,
                                                           temp=temp,
                                                           desired_temp=desired_T,
                                                           box_size=size,
                                                           relax_steps=relax_steps)
```

Analysis of solid, liquid and gas phases

```python
liquid_pc=liquid_observables[2]
solid_pc=solid_observables[2]
gas_pc=gas_observables[2]
```

```python
def parabola(x,a,b):
    return a+b*x**2

def line(x,a,b):
    return a+b*x
```

```python
y=gas_dif[0][50:175]
x=np.array(range(len(x)))
params, errors = curve_fit(parabola,x, y)
```
