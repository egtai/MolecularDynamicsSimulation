```python
import simulation as sk
import visualization as vs
import matplotlib.pyplot as plt
import numpy as np
import tkinter as tk
```

```python
def verlet_simulation(step, steps, sigma, epsilon, mass, temp, desired_temp, box_size, relax_steps):
    dm_time = step * (epsilon/(mass*sigma**2))**(1/2)
    square = np.array([[0.1,0.5,0],[0.9,0.5,0]])
    vels = sk.init_velocity(len(square), temp, mass)
    return sk.vervlet_evolution(pos=square, 
                                vel=vels, 
                                time_step=dm_time, 
                                steps=steps, 
                                mass=mass, 
                                sigma=sigma, 
                                epsilon=epsilon, 
                                size=box_size, 
                                temp=temp,
                                relaxation_steps=relax_steps, 
                                desired_temp=desired_temp)
```

```python
sigma, epsilon, mass = sk.argon_parameters()
t_step = 1e-15
temp = 100
steps = 10000
relax_steps = 10
size = 1.5
desired_T = 10
data, energy, observables = verlet_simulation(step=t_step,
                                                           steps=steps,
                                                           sigma=sigma,
                                                           epsilon=epsilon,
                                                           mass=mass,
                                                           temp=temp,
                                                           desired_temp=desired_T,
                                                           box_size=size,
                                                           relax_steps=relax_steps)
```

```python
fig = plt.figure()
ax4 = fig.add_subplot()
ax4.plot(range(steps), np.sum(energy[0].T,axis=0), lw=2)
ax4.plot(range(steps), np.sum(energy[1].T,axis=0), lw=2)
ax4.plot(range(steps),np.sum(energy[0].T+energy[1].T,axis=0), lw=2)
ax4.set_title("Two particles: energy")
ax4.grid(True)
ax4.set_xlim(1100,4500)
ax4.set_ylim(-5,25)
ax4.set_xlabel(r"$t/t^*$")
ax4.set_ylabel(r"$E/\epsilon^*$")
ax4.legend(['Kinetic','Potential','Total'])
plt.show()
```

Lennard Jones potential analysis: 
* Force in dimensionless units.
* Check magnitude of the force for two particles in 3D at a distance 2

```python
x = np.arange(0.25,5,0.01)
y = -sk.dlennard_jones(x)
z = sk.lennard_jones(x)
```

```python
%matplotlib inline
plt.ylim(-5,5)
plt.xlabel(f"$|r_1-r_2|/\sigma$")
plt.ylabel(f"$U(r)/\epsilon$")
plt.title("Lennard-Jones")
plt.plot(x, z)
plt.plot(x, y)
plt.legend(["LJ potential","LJ force"])
```
